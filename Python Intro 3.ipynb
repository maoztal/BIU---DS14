# Working with Pandas Data Frames

```python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')
#iris = pd.read_csv("iris.csv")
iris.head()
```

```python
iris.tail()
```

```python
## add new colums to the df
iris['sepal_ratio'] = iris.sepal_length / iris.sepal_width
iris.head()
```

```python
iris['petal_ratio'] = iris.petal_length / iris['petal_width']
iris.head()
```

```python
## now proceed with checking the values of the df
iris.describe()
```

```python
import numpy as np
## create categorical variables for the sepal and petal width
iris['petalw_cat'] = np.floor(iris.petal_width)
iris['sepalw_cat'] = np.floor(iris.sepal_width)
iris.describe()
```

```python
## subsetting
iris[iris.petalw_cat==2]
```

```python
iris.describe()
```

```python
### subsetting for given values
iris[iris.species.isin(['virginica','versicolor'])]
```

```python
# subsetting between values

iris2 = iris.copy()
iris2 = iris2[iris2['petal_length'].between(5,6)]
iris2.describe()
```

# Joining dataframes
 
The merge function help us to make the joins. The function parameters are the following;

<code> pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)
</code>
  
left and right are the tables.
 
how= is the type of join to be used. It can be "inner", "left", "right", and "outer". The default is "inner".
 
on= is the column we will use to make the join. In case the name is different in both tables, we will define them with "left_on=" and "right_on=" and will give the respective column names in each table.
 
Let's try to join the classroom and students tables using inner join.



```python
class_student = pd.merge(classrooms, students, how='inner', on='StudentId')
```

```python
class_course = pd.merge(classrooms, courses, how='left', on='CourseId')
```
